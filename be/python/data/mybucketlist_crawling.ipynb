```python
bucketlist_titles = []
bucketlist_categories = []
```

```python
import selenium
from selenium import webdriver
import time
from selenium.webdriver.common.by import By

WEBDRIVER_OPTIONS = webdriver.ChromeOptions()

# 창을 띄우지 않는 옵션
# 실제 크롤링할 때 활성화
# WEBDRIVER_OPTIONS.add_arguments("headless")

webDriver = webdriver.Chrome(options=WEBDRIVER_OPTIONS)

url = ["https://findmybucketlist.com/search/newresult?category=bucket&params=%3Ffield%3Dbd_category%26value%3DNATURE"]

category = ['NATURE', 'LIFESTYLE', 'SHOPPING', 'TRAVLE', 'CULTURE','NEW%2520SKILL', 'FOOD', 'OUTDOOR', 'SPORTS']

# webDriver.get(url + category[])
# time.sleep(1)

bucketlist_list = []
   
```

```python
for cate in category:
    print(cate)
    webDriver.get(url + cate)
    time.sleep(3)
    
    titles = []
    categories = []
    
    # 팝업창 닫기
    if cate == 'NATURE':
        popup = webDriver.find_element(By.CLASS_NAME, 'popup-today-text')
        popup.click()
        time.sleep(3)
    
    SCROLL_PAUSE_TIME = 5
    last_height = webDriver.execute_script("return document.body.scrollHeight")
    while True:
        webDriver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)
        new_height = webDriver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        time.sleep(SCROLL_PAUSE_TIME)

    titles = webDriver.find_elements(By.CLASS_NAME, 'results--box-desc__title-bucket')
    categories = webDriver.find_elements(By.CLASS_NAME, 'results--box-desc__bd_category_name')

    print(len(titles))
    
    for title in titles:
        bucketlist_titles.append(title.text)

    for category in categories:
        bucketlist_categories.append(category.text)
        
#     logo = webDriver.find_element(By.CLASS_NAME, 'header__logo-holder')
#     logo.click()
#     time.sleep(3)
```

```python
## 한 url에서만 크롤링
webDriver.get('https://findmybucketlist.com/search/newresult?category=bucket&params=%3Ffield%3Dbd_category%26value%3DNEW%2520SKILL')
time.sleep(3)

# 팝업창 닫기
popup = webDriver.find_element(By.CLASS_NAME, 'popup-today-text')
popup.click()
time.sleep(3)

SCROLL_PAUSE_TIME = 3
last_height = webDriver.execute_script("return document.body.scrollHeight")
while True:
    
    webDriver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = webDriver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    time.sleep(SCROLL_PAUSE_TIME)

titles = webDriver.find_elements(By.CLASS_NAME, 'results--box-desc__title-bucket')
categories = webDriver.find_elements(By.CLASS_NAME, 'results--box-desc__bd_category_name')

print(len(titles))

for title in titles:
    bucketlist_titles.append(title.text)

for category in categories:
    bucketlist_categories.append(category.text)

#     logo = webDriver.find_element(By.CLASS_NAME, 'header__logo-holder')
#     logo.click()
#     time.sleep(3)
```

```python
# 파일에 처음 쓸 때
import csv
with open('./mybucketlist_all2.csv', mode="w", newline="", encoding="cp949") as bucketlist_file:
    writer = csv.writer(bucketlist_file)
    
    header = ['title', 'category']
    writer.writerow(header)
    
    for bucketlist in zip(bucketlist_titles, bucketlist_categories):
        
        writer.writerow([bucketlist[0], bucketlist[1]])
        
webDriver.quit()
```

```python
# 파일에 이어 쓸 때
import csv
with open('./mybucketlist_all2.csv', mode="a", newline="", encoding="cp949") as bucketlist_file:
    writer = csv.writer(bucketlist_file)
    
    for bucketlist in zip(bucketlist_titles, bucketlist_categories):
        
        writer.writerow([bucketlist[0], bucketlist[1]])
        
webDriver.quit()
```
