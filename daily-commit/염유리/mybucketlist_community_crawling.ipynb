```python
webDriver = webdriver.Chrome(options=WEBDRIVER_OPTIONS)
```

```python
url = "http://mybucketlist.co.kr/"
```

```python
import time
webDriver.get(url)
time.sleep(1)
```

```python
from selenium.webdriver.common.by import By
bucketlist_list = []
# elements = webDriver.find_elements(By.CLASS_NAME, 'post-title')
# bucketlist_list.append(elements)
```

```python
# range 30까지 했을때 중복제거 후 106개
# range 100까지 했을때 중복제거 후 127개
# range 200까지 했을때 중복제거 후 142개

for page in range(1, 200):
#     if page == 1:
#         pass
#     else:
    print(page)
    SCROLL_PAUSE_TIME = 5
    last_height = webDriver.execute_script("return document.body.scrollHeight")
    while True:
        webDriver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)
        new_height = webDriver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        
    elements = webDriver.find_elements(By.CLASS_NAME, 'post-title')
    
    for element in elements:
        bucketlist_list.append( element.get_attribute("data-title"))
    
```

```python
len(bucketlist_list)
len(set(bucketlist_list))

bucketlist_list = set(bucketlist_list)
```

```python
bucketlist_list
```

```python
import csv
with open('./bucketlist_community_range200.csv', mode="w", newline="") as bucketlist_file:
    writer = csv.writer(bucketlist_file)
    
    header = ['title']
    writer.writerow(header)
    
    for bucketlist in bucketlist_list:
        
        writer.writerow([bucketlist])

webDriver.quit()
```
